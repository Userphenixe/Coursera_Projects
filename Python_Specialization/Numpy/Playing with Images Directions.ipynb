```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
jellyfish = img.imread("jellyfish.png")
plt.imshow(jellyfish, cmap= 'gray')
```

```python
np.shape(jellyfish)
```

```python
jellyfish_copy = jellyfish.copy()
```

```python
lighter_image = np.clip(jellyfish * 1.5, 0, 1)
plt.imshow(lighter_image, cmap='gray', vmin=0, vmax=1)
```

```python
jellyfish_copy.transpose()
jellyfish_copy
```

```python
lighter_image_copy = np.clip(jellyfish_copy * 1.5, 0, 1)
plt.imshow(lighter_image_copy, cmap='gray', vmin=0, vmax=1)
```
